```python
# Task 1.2.3 Clean DF1
# Remove "NAN" value
df1.dropna(inplace=True)
```

```python
df1['PRICE_USD'].head()
```

```python
# This will remove the dollar sign
df1['PRICE_USD'].str.replace("$", "", regex=False).head()
```

```python
# This remove both the dollar sign and the comma
df1['PRICE_USD'].str.replace("$", "", regex=False).str.replace(",", "").head()
```

```python
# This remove both the dollar sign,the comma and change the data type to float
df1['PRICE_USD'].str.replace("$", "", regex=False).str.replace(",", "").astype(float).head()
```

```python
# Transform the price from object to float
df1['PRICE_USD'] = (
    df1['PRICE_USD'].str.replace("$", "", regex=False)
    .str.replace(",", "")
    .astype(float).head()
)
```
